```python
import requests
import pandas as pd
import time
import datetime

# Store all collected data
all_data = []

# Adjust this (15–20 is ideal)
NUM_CALLS = 15

for i in range(NUM_CALLS):
    print(f"Fetching batch {i+1}/{NUM_CALLS}...")

    try:
        url = "https://opensky-network.org/api/states/all"
        response = requests.get(url)

        # Check if request successful
        if response.status_code != 200:
            print("Error:", response.status_code)
            continue

        data = response.json()
        states = data.get("states", [])

        # Capture timestamp for this batch
        current_time = datetime.datetime.now()

        for row in states:
            # Append row + timestamp
            all_data.append(row + [current_time])

        print(f"Collected {len(states)} rows")

    except Exception as e:
        print("Error occurred:", e)

    # Delay to avoid rate limiting
    time.sleep(10)

# Column names (OpenSky + timestamp)
columns = [
    "icao24", "callsign", "origin_country", "time_position",
    "last_contact", "longitude", "latitude", "baro_altitude",
    "on_ground", "velocity", "heading", "vertical_rate",
    "sensors", "geo_altitude", "squawk", "spi", "position_source",
    "collection_time"
]

# Create DataFrame
df = pd.DataFrame(all_data, columns=columns)

# Save RAW dataset (important)
df.to_csv("opensky.csv", index=False)

print("\nData collection complete!")
print("Total rows collected:", len(df))
```
